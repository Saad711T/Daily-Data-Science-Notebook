# Data structures with Python

### Types of DS in Python :
- Lists
- Tuples
- Sets
- Stacks
- Queues
- Trees

### Lists - Tuples - Sets

#### Lists :
- Ordered
- Allows duplicates
- Store different data types

```python
list = [1,"two",3,4.0]

# First element have zero index
```

```python
print(list[0])
```

#### Sets :
- Unordered
- Unique elements
- Store different data types

```python
set = {1,2.0,"three",4}

# Cannot add another 1 - 2.0 - "three" string or 4 elements in same set
```

#### Tuples :
- Ordered
- Elements cannot be modified
- Allows duplicates
- Store different data types

It's like upgraded lists

```python
tuple = (1,2,3.333333,True)
```

Stacks , Queues and Trees just combined with those primary data structures in Python.

### Operations

We can do many operations in data structures like : add , remove and append...

#### Append

```python
list.append(4)

print(list)
```

#### Add

```python
set.add(10)

print(set)
```

#### Remove

```python
set.remove("three")

print(set)
```
