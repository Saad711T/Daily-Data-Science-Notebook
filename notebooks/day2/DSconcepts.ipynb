# Data science concepts and articles

DS or Data Science is a field combining mathematics and statistics , artificial intelligance , and computer science articles to cleaning and analyzing data for decision making process.

## Main articles

- Applied Mathematics (Statistics and Probability - Physics - ML Math) : Used in Many Scientific Computing applications around the world along side with Data analysis and machine learning.
- Pure Mathematics (Linear Algebra - Calculus - Operations Research) : Used mainly in ML configuration and subsystems.
- Data Analysis : Cleaning , Extracting and Analyzing data to Decision making with skills like web-scraping , Dashboards making.
- Programming (Python-R-Scala) : Programming is the most important thing in data science for using tools and programming libraries and frameworks such as : NumPy - Pandas and others...
- Databases (SQL) : Databases used primarily to storing big data using strong query language such as SQL or Structured Query Language.
- Machine learning : To create models that predict the results.
