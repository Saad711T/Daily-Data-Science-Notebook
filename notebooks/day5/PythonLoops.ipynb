# Python Loops

```python
fruits = ["apple", "banana", "watermelon"]
```

### For loop :

```python
for fruit in fruits:
    print(fruit)
```

#### Using for loop to print numbers from 0 to 9 (10 numbers) :

```python
for i in range(10):
    print(i)
```

#### Using for loop within numbers range :

```python
for n in range(1, 5):
  print(n) 
```

#### Using else with loops :

```python
for x in range(5):
  print(x)

  
else:
  print("Finish")
```

```python
for x in [0, 1, 2]:
  pass
```

### While loop

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

#### While loop with Break condition

```python
i = 1
while i < 5:
  print(i)
  if i == 3:
    break
  i += 1
```

#### While loop with Break condition

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```
